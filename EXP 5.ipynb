```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

Line chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

BAR CHART

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

RANDOM BAR CHART

```python
x = np.arange(1, 11)                 # x values (1 to 10)
y = np.random.randint(1, 100, 10)    # random integers for y

# Create bar chart
plt.bar(x, y, color='pink', edgecolor='black')

# Add labels and title
plt.title("Random Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis (Random Values)")

# Show the chart
plt.show()
```

```python
a=(1,4,3,5,6,8)
b=(2,5,6,9,3,1)
plt.scatter(a,b)
plt.title("scatter plot")
plt.scatter(a, b, color='pink', edgecolor='purple')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
